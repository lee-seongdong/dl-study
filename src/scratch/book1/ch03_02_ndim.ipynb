```python
import numpy as np
```

```python
A = np.array([1, 2, 3, 4, 5])
print('A:\n', A)
print('A.ndim: ', A.ndim)
print('A.shape: ', A.shape)
print('A.dtype: ', A.dtype)
print()

B = np.array([[1, 2], [3, 4], [5, 6]])
print('B:\n', B)
print('B.ndim: ', B.ndim)
print('B.shape: ', B.shape)
print('B.dtype: ', B.dtype)

```

```python
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])

print(np.dot(A, B))
```

```python
A = np.array([[1, 2, 3], [4, 5, 6]])
B = np.array([[1, 2], [3, 4], [5, 6]])
C = np.dot(A, B)

# 행렬곱
# dim(a, b) * dim(c, d) = dim(a, d). 단 b == c여야함
# ex) dim(2, 3) * dim(3, 2) = dim(2, 2)

print('A.shape: ', A.shape)
print('A:\n', A, '\n')
print('B.shape: ', B.shape)
print('B:\n', B, '\n')
print('C.shape: ', C.shape)
print('C:\n', C, '\n')
```
