```python
import torch
import sentencepiece as spm

corpus_dir = "C:/Users/USER/Korpora/kowikitext/"
vocab_file = "kowiki_32000.model"
vocab = spm.SentencePieceProcessor()
vocab.load(corpus_dir + vocab_file)

lines = [
    "감기 조심하세요.",
    "테스트 입니다?",
]

inputs = []
for line in lines:
    pieces = vocab.EncodeAsPieces(line)
    ids = vocab.EncodeAsIds(line)
    inputs.append(torch.tensor(ids))
    print(pieces)

inputs = torch.nn.utils.rnn.pad_sequence(inputs, batch_first=True, padding_value=0)
print(inputs.size())
print(inputs)
```

```python
# transformer의 embedding은 input embedding과 positin embedding을 함께 사용함

# input embedding
n_vocab = len(vocab)
d_hidden = 128
nn_emb = torch.nn.Embedding(n_vocab, d_hidden)

input_embeddings = nn_emb(inputs)
print(input_embeddings.size()) # [input_size, max_token_size, hidden_size]

# position embedding
```
