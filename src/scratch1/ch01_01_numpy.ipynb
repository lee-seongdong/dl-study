```python
import numpy as np
```

```python
x = np.array([1, 2, 3])
print(x)
print(type(x))
```

```python
# numpy 산술 연산
x = np.array([1.0, 2.0, 3.0])
y = np.array([4., 5., 6.])
print(x + y)
print(x - y)
print(x * y)
print(x / y)
```

```python
x = np.array([1, 2, 3])
print(x / 2)
```

```python
# numpy 다차원 배열
A = np.array([[1, 2], [3, 4]])
print(A)
print(A.shape)
print(A.dtype)
```

```python
# numpy 브로드캐스트
A = np.array([[1, 2], [3, 4]])
print(A * 3)
B = np.array([10, 20])
print(A + B)
print(A * B)
```

```python
# numpy 원소 접근
X = np.array([[51, 55], [14, 19], [0, 4]])
print(X)
print(X[0])
print(X[0][1])
```

```python
X = np.array([[51, 55], [14, 19], [0, 4]])
for row in X:
    print(row)
```

```python
X = np.array([[51, 55], [14, 19], [0, 4]])
X = X.flatten()
print(X)

print(X[np.array([0, 2, 4])]) # 인덱스가 0, 2, 4인 원소 추출
```

```python
X = np.array([[51, 55], [14, 19], [0, 4]])
X = X.flatten()
print(X > 15) # 15보다 큰 원소를 추출
print(X[X > 15])
```
